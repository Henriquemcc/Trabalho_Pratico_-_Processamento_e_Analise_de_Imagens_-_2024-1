# Como converter uma matriz RGB em uma matriz de tons de cinza

```python
def rgb_to_grayscale(rgb_image):
    # Usar a fórmula da luminância para converter para tons de cinza
    # grayscale = 0.2989 * R + 0.5870 * G + 0.1140 * B
    grayscale_image = numpy.dot(rgb_image[...,:3], [0.2989, 0.5870, 0.1140])
    return grayscale_image
```

```python
imagem_convertida = rgb_to_grayscale(matriz_rgb)
```

```python
type(imagem_convertida)
```

```python
grayscale_image_pil = PIL.Image.fromarray(imagem_convertida.astype(numpy.uint8))
```

```python
grayscale_image_pil.save('imagem_tons_de_cinza.jpg')
```
